```python
import time
start = time.time()


# Read from bronze table
silver_df = spark.table('project_catalog.medallion_project.bronze_table')\
    .dropDuplicates()

# Save as silver table
silver_df.write.format('delta')\
    .mode('overwrite')\
    .saveAsTable('project_catalog.medallion_project.silver_table')
```

```python
spark.sql("OPTIMIZE project_catalog.medallion_project.silver_table ZORDER BY (Region, Product_Category)")
```
